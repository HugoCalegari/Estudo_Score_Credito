# Leitura da base de dados de treino

```python
treino = pd.read_parquet('../1.Base_de_Dados/Dados_Bronze/treino.parquet')
treino.head()
```

```python
treino.info()
```

```python
def retira_underscore(df, variavel):
    
    indice = df[df[variavel].str.contains('_')].index
    df.loc[indice,variavel] = df.loc[indice,variavel].str.replace('_', '')

    return(df)
```

```python
retira_underscore(treino, 'Age')['Age'].astype('int').describe()
```
